```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 4.5, 5, 6, ]
y = [0, 4, 9, 16, -1, 0.5, 1.5,]

# if y < 0 print red else print blue
colors = ['red' if y_ <= 0 else 'blue' for y_ in y]
plt.scatter(x, y, c=colors)
# if point is red print red line else green
plt.plot(x, y, c='black')
plt.grid()
plt.show()
```

```python
x = np.linspace(-15, 10, 150)
# y = x**2
y = np.sin(x)
plt.grid()
plt.plot(x, y)
```

```python
def calcula_y(x):
    return x**2 + 7*x - 3
```

```python
y = calcula_y(x)
# añadir una rejilla al plot
plt.grid()
plt.plot(x, y)
```

```python
# calcula la derivada de cualquier función
def derivada(f, x, incremento=0.01):
    return (f(x + incremento) - f(x)) / incremento

def derivada_concreta(x):
    # y' = dy/dx = 2*x +7
    return 2*x + 7

derivada1 = derivada(calcula_y, x,4)
derivada2 = derivada_concreta(x)

plt.grid()
plt.plot(x, y)
plt.plot(x, derivada1)
plt.plot(x, derivada2)
plt.show()
```

```python
xinicial = -15
yinicial = calcula_y(xinicial)
derivada3 = derivada(calcula_y, xinicial, 0.01)
print("---Paso inicial---")
print(f'x inicial: {xinicial}')
print(f'y inicial: {yinicial}')
print(f'pendiente: {derivada3}')
```

```python
paso = 0.0001
x2 = xinicial
i = 0
while(True):
    x2 += paso
    y2 = calcula_y(x2)
    derivada4 = derivada(calcula_y, x2, paso)
    i+=1
    if derivada4 < paso and derivada4 > -paso:
        break
print(f"---Paso {i}---")
print(f'x: {x2}')
print(f'y: {y2}')
print(f'pendiente: {derivada4}')
plt.grid()
plt.plot(x, y)
plt.plot(x, derivada1)
plt.show()
```
