```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 4.5, 5, 6, ]
y = [0, 4, 9, 16, -1, 0.5, 1.5,]

# if y < 0 print red else print blue
colors = ['red' if y_ <= 0 else 'blue' for y_ in y]
plt.scatter(x, y, c=colors)
# if point is red print red line else green
plt.plot(x, y, c='black')
plt.grid()
plt.show()
```

```python
x = np.linspace(-15, 10, 150)
# y = x**2
y = np.sin(x)
plt.grid()
plt.plot(x, y)
```

```python
def calcula_y(x):
    return x**2 + 7*x - 3
```

```python
y = calcula_y(x)
# aadir una rejilla al plot
plt.grid()
plt.plot(x, y)
```

```python
# calcula la derivada de la funcin
def derivada(f, x, h=0.01):
    return (f(x + h) - f(x)) / h

derivada1 = derivada(calcula_y, -3.505)
print(derivada1)
print(calcula_y(-3.505))
```
